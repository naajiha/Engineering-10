Then I will take a part of the scipy integration module and only take solve_ivp because I install the whole thing the volume of tools is more than needed in scipy.

```python
import numpy as np
from scipy.integrate import solve_ivp
```

I plan to define a function named func_int to represent the function that will be integrated with t(time) and P(population) being my parameters. Then within the function I will define variables K, r, and T and assign them the respective values of 1,000,000, .00231, and 1000. Then I will return the right hand side of the equation.

```python
#time is t
#population P
#K = 1,000,000 cells
#r = 0.00231/growth period
#T = 1000 cells
def func_int(t, P):
    K = 1000000
    r = .00231
    T = 1000
    return -r * P[0] * (1-P[0]/K)*(1-P[0]/T)
```

I will run the solver and save results in result variable. I will set starting population value to be 500000 as that is half of K. I will the use a guess and check model to figure out the time it takes for the population to reach 99% carrying capacity.

```python
result = solve_ivp(func_int, (0, 4), [500000])
print(result.y)
```
