I have been asked to work at engineering construction company(ECC) with national home construction company(NHCC) as my client. ECC wants to work with NHCC to build driveways for Home Constructors across the country. My company ECC cares about the environment so we prefer concrete driveways to asphalt driveways. The concrete driveways should be used because concrete is more envionmentally friendly, durable, longer lasting and the most customizable in comparison to asphalt. The driveway must have the dimensions 30 feet by 20 feet by 4 inches(.33 feet). The concrete driveway will be made out of the standard basic ingredients, aggregate, sand, gravel, cement. There will be no premade concrete mix as my boss has requested me to plan for usage of standard ingredients.

For this project, I will first need to research the pros and cons of concrete and see what meterials are needed to make concrete. I also need to research whether or not I will be using premade concrete or concrete made from basic ingredients. Then I need to calculate how much of each I would need. Then I will need to research the cost of each of the ingredients for premade concrete and concrete from basic ingredients. Then depending on all these factors I will decide which one is the most cost effective and efficient concrete decision to make.

concrete is more affordable in comparison to asphalt. It is also longer lasting allowing homeowners not to worry about the state of their driveway for 50+ years. Additionally, concrete can survive under harsh whether conditions and it tenacious through high stress loads. And concrete is very eco-friendly as it is 100% reusable. Finally, it is incredibly custimizable to a client's needs.



```python
#in the next three lines I am assigning variables to the dimensions given for the concrete driveway 
thick_inch = 4
long_ft = 30
wide_ft = 20

#I converted the thickness value from inches to feet to have the same units of volume for all the dimensions
thick_ft = thick_inch / 12

''''In the next two lines I am finding the volume of concrete I will need for the entire driveway by multiplying all
given dimensions together then I added a print statement so my company(Engineering Construction Company) and our client 
(National Home Construction Company) will know the final calculation'''

concrete_CubicFt = long_ft * wide_ft * thick_ft
print(concrete_CubicFt, 'feet cubed of concrete is needed')
```

```python
#newest version

#here I am defining the variable for how much a single bag of cement weighs and how many cubic feet the bag fills
portCementWeight_lbs = 94
portlandConc_ft3 = 4.5
concrete_CubicFt = 200

'''here I am finding how many bags of cement in total I will need to finish the 200 cubic feet driveway. then in 
next line I am seeing how many lbs all the cement bags will be'''
cement_BagTotal = concrete_CubicFt / portlandConc_ft3
Totallbs = cement_BagTotal * portCementWeight_lbs

#here I am finding how many lbs of cement I will need with the added 10%
PCementTotal_lbs = (Totallbs) * 1.1
print(round(PCementTotal_lbs), 'lbs cement total')

'''here I am finding out how many lbs total I need of sand by taking the cement in multiplying in by 2,
following the ratio given'''
SandTotal_lbs = (Totallbs * 2) * 1.1
print(round(SandTotal_lbs), 'lbs sand total')

'''here I am finding out how many lbs total I need of gravel by taking the cement in multiplying in by 3,
following the ratio given'''
GravelTotal_lbs = (Totallbs * 3) * 1.1
print(round(GravelTotal_lbs), 'lbs gravel total')
```

```python
#cement = $13.37/94 lb bag
#sand = $3.97/50 lb bag
#gravel = $5.37/50 lb bag 

'''here I am finding how many bags of cement I will need by taking total weight of cement needed divided by one bag weight
then I take how many bags of cement I will need (cementBagTotal) and multiply it by the price of one bag (found through 
online source)'''
cementBagTotal = PCementTotal_lbs / 94
cementCostTotal = 13.37 * cementBagTotal
print(int(cementBagTotal + 1), 'bags cement', '$', (round(cementCostTotal, 2))) 

#here I did the same thing as I did for cement
sandBagTotal = SandTotal_lbs / 50
sandCostTotal = 3.97 * sandBagTotal
print(int(sandBagTotal + 1), 'bags sand', '$', (round(sandCostTotal, 2)) )

#here I did the same thing as I did for cement
gravelBagTotal = GravelTotal_lbs / 50
gravelCostTotal = 5.37 * gravelBagTotal
print(int(gravelBagTotal + 1), 'bags gravel', '$', (round(gravelCostTotal, 2)))

#in the next lines I simply added all ingredients to get total final cost, then gave a print statement so employer can see
totalCost = cementCostTotal + sandCostTotal + gravelCostTotal
print('$', (round(totalCost, 2)), 'is the total cost')
```

```python
#here I am caluclating the volume of how much concrete one bag of premix concrete makes
bagVolume = 2 * 2*  (4/12)

#here I am calculating how many bags are needed total to have enough concrete to make 200 cubic feet of concrete
premix_bagTotal = concrete_CubicFt / bagVolume
print(premix_bagTotal, 'bags of premixed concrete are needed')

#each bag of premix concrete (80 lb) is $5.17
premixTotalCost = premix_bagTotal * 5.17
print('$', premixTotalCost, 'is the total cost to do the job')

```

I recommend to use the concrete from basic ingredients, this will allow to save $1042.56 in comparison to the premix concrete. With the basic ingredients concrete, the saved money that would have been used to buy the premix concrete can be used towards the labor, tax, benefit costs etc.
