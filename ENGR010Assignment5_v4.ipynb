First I will create an empty list for air liquid and air gas. Then I will create a with statement that opens the spreadsheet csv file I make and reads it. Then I will store  the liquid string under flag to mark it. Then I will create a for loop to go through the elements in the list of the csv file. I will append the liquid and gas lists. Then I will print both lists.

```python
air_liquid = [] #creating empty list
air_gas = [] #creating empty list

with open('test_data.csv', 'r') as fh: #creating with statement to read file data
    my_data = fh.readline()[2:]
    flag = 'Liquid'
    for line in fh:
        line = line.rstrip()
        row = line.split(',')
        if row[0] == 'Gas':
            flag == 'Gas'
        elif flag == 'Liquid':
            air_liquid.append(row)
        else:
            air_gas.append(row)
    print(air_liquid)
    print(air_gas)
 
```

First I will create a function with one parameter to represent data. Then  will create an empty density dictionary. I will make a for loop that iterates though the table data list. Then in the list it will loop and be given a number representing which value in the list it is taking as the key and definition. Then the dictionary will be returned.

```python
def new_dict(air_data):
    density_dictionary = {}
    for i in range(0, len(air_data)):
        density_dictionary.update({air_data[i][0] : air_data[i][3]})
    return density_dictionary
new_dict(air_liquid)
```

I will create another function with the same parameter. Then I will create a dictionary for each element(pressure, density, cp, and dv The I will create a dictionary for the overall liquid so I can print the values (type and desired value) together. Then within the for loop I will make I will update each of these dictionaries with the corresponding key and definition according to my table. Then I will update the liquid dictionary such that the pressure, cp, and dv, are all callable values. Then they can be called on.

```python
def create_all_dicts(air_liquid):
    pressure_dict = {}
    cp_dict = {}
    dv_dict = {}
    liquid_dict = {}
    for i in range(0, len(air_liquid)):
        pressure_dict.update({air_liquid[i][0] : air_liquid[i][2]})
        cp_dict.update({air_liquid[i][0] : air_liquid[i][5]})
        dv_dict.update({air_liquid[i][0] : air_liquid[i][8]})
    liquid_dict.update({'pressure' : pressure_dict, 'cp' : cp_dict, 'dv' : dv_dict})
    return liquid_dict

liquid_dict = create_all_dicts(air_liquid)
```

```python
print(liquid_dict['pressure']['100'])
print(liquid_dict['cp']['120'])
print(liquid_dict['dv']['65'])
```
