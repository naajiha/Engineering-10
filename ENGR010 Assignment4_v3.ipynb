**3. A Single Conversion Function for lengths, areas and volumes**

I plan to define a function with three parameters, measure type(length, area, volume), direction(inches to meters, meters to inches, inches squared to meters squared, meters squared to inches squared, etc.), or volume conversions(inches cubed to meters cubed, meters cubed to inches cubed, etc.). Then I plan to create an if statement nested within an if statement, the first if statement determining measure type, the second if statements determining between meters to inches or inches to meters, or meters squared/cubed to inches squared/cubed, to inches squared/cubed to meters squared/cubed. Then, after that is determined the function has been programmed to apply the appropriate math to the number entered.

```python
#  Your function that can convert SI length, area or volume units to EEU units and vice versa
def twowayconversion(measure, direction, number):  
    if measure == 'length':
        if direction == 'meters_2_inches':
            convertedNumber = 3.281 * 12 * number, 'inches'
            
        elif direction == 'inches_2_meters':
            convertedNumber = (number / (3.281 * 12)), 'meters'
    if measure == 'area':
        if direction == 'meterssquared_2_inchessquared':
            convertedNumber = number * 1550, 'inches squared'
            
        elif direction == 'inchessquared_2_meterssquared':
            convertedNumber = number / 1550, 'meters squared'
            
    elif measure == 'volume':
        if direction == 'inchescubed_2_meterscubed':
            convertedNumber = number / 61024, 'meters cubed'
          
        elif direction == 'meterscubed_2_inchescubed':
            convertedNumber = number * 61024, 'inches cubed'
    return convertedNumber        
```

**4. A Single Conversion Function for everything**

I plan to do this in the same way that step 3 was completed, I will simply add another if statement with a nested if statement in order to add temperature to the decision the program decides on whether to convert an entered value and if temperature is chosen then the appropriate conversions from farenheit to celsius or celsius to farenheit will be made. 

```python
#  Your function that can convert any of the required SI quantities into EEU units or vice versa
def twowayconversion(measure, direction, number):  
    if measure == 'length':
        if direction == 'meters_2_inches':
            convertedNumber = 3.281 * 12 * number, 'inches'
            
        elif direction == 'inches_2_meters':
            convertedNumber = (number / (3.281 * 12)), 'meters'
    if measure == 'area':
        if direction == 'meterssquared_2_inchessquared':
            convertedNumber = number * 1550, 'inches squared'
            
        elif direction == 'inchessquared_2_meterssquared':
            convertedNumber = number / 1550, 'meters squared'
            
    if measure == 'volume':
        if direction == 'inchescubed_2_meterscubed':
            convertedNumber = number / 61024, 'meterscubed'
          
        elif direction == 'meterscubed_2_inchescubed':
            convertedNumber = number * 61024, 'inchescubed'
    elif measure == 'temperature':
        if direction == 'celsius_2_farenheit':
            convertedNumber = (number * (9 / 5)) + 32, 'degrees farenheit'
        if direction == 'farenheit_2_celsius':
            convertedNumber = (number - 32) * (5 / 9), 'degrees celsius'
    return convertedNumber
            
    
        
        
```

**5. Convert a whole list of values**

My boss want me to convert a range of numbers from one value to another set of values(example, meters to inches. 
I plan to do this by creating a function called conversion with two parameters, direction and list of numbers. Then i will create an empty dictionary I can store terms into. Then I will create a series of if statements ready to take and conversion(meters to inches, inches to meters, etc). Within this there will be a dictionary.update method in order to update the dictionary with the elements from another dictionary object. Then I will return dictionary so the function can be called on when printing.

```python
def conversion(direction, numberList): 
    dictionary = {}
    for number in numberList:
        if direction == 'meters_2_inches':
            dictionary.update({number:3.281*12*number})
        elif direction == 'inches_2_meters':
            dictionary.update({number: number/(3.281*12)})
        elif direction == 'meterssquared_2_inchessquared':
            dictionary.update({number : number*1550})
        elif direction == 'inchessquared_2_meterssquared':
            dictionary.update({number : number / 1550})
        elif direction == 'inchescubed_2_meterscubed':
            dictionary.update({number : number / 61024})
        elif direction == 'meterscubed_2_inchescubed':
            dictionary.update({number : number * 61024})
        elif direction == 'celsius_2_farenheit':
            dictionary.update({number : (number * (9 / 5)) + 32})
        elif direction == 'farenheit_2_celsius':
            dictionary.update({number : (number - 32) * (5 / 9)})
    return dictionary
```

```python
# Test your code by converting this list of lengths from inches to meters: [1., 12.0, 39.37, 144., 5280.*12.]
# Print the dictionary keys and values one at a time in a table with headings for inches and meters.
dictionary_results = conversion('inches_2_meters', [1., 12.0, 39.37, 144., 5280.*12.] )
print('inches\t\tmeters')
for value in dictionary_results:
    print((str(value)) + '\t\t' + (str(dictionary_results[value])))
```
