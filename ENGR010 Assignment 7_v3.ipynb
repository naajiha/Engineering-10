<strong>1. Reading in the data</strong>

I plan to complete this section by first importing the pandas and numpy extensions to access their functions and tools. Then I will initialize my numpy array using the np.zeroes function. Then I will read the machine file using the pandas command.
Then I will create 3 nested for loops to read out machines, month and trial. The information will then be stored in my created array.

```python
import pandas as pd
import numpy as np
array = np.zeros((5, 3, 20))
index = 0
df = pd.read_csv('machine.csv')
for i in range(5):
    for j in range(3):
        for k in range(20):
            array[i][j][k] = df.loc[index, 'Length']
print(array)
```

<strong>2. Plot the first machine (Machine 0) data</strong>

First I will import the matplotlib library to plot graphs. Then I will create the first month array plot. Then I will plot this with the plt.plot function.

```python
import matplotlib.pyplot as plt
first_month_rosita = array[0][0]
plt.plot(first_month_rosita, 'ro')
```
